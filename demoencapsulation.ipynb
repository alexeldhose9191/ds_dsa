```python
class Cart:
    product = "Laptop"
    price = 36000
    quantity = 2
    total = 0

c = Cart()

c.total = c.price * c.quantity

print("Product :", c.product)
print("Price   :", c.price)
print("Quantity:", c.quantity)
print("Total Amount to Pay:", c.total)
```
